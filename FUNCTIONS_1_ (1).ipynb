1) Write a Python function to sum all the numbers in a list.

Sample List : (8, 2, 3, 0, 7)

Expected Output : 20


```python
#WRITE YOUR CODE HERE
def sum_list(numbers):
    return sum(numbers)
sample_list = [8, 2, 3, 0, 7]
print("Sum of the list:", sum_list(sample_list))
```

2) Write a Python function to multiply all the numbers in a list.

Sample List : (8, 2, 3, -1, 7)

Expected Output : -336

```python
#WRITE YOUR CODE HERE
def multiply_list(numbers):
    result = 1
    for num in numbers:
        result *= num
    return result
sample_list = [8, 2, 3, -1, 7]
print("Product of the list:", multiply_list(sample_list))
```

3) Write a Python program to reverse a string.

Sample String : "1234abcd"

Expected Output : "dcba4321"

```python
#WRITE YOUR CODE HERE
sample_string = "1234abcd"
reversed_string = sample_string[::-1]
print("Reversed String:", reversed_string)
```

4) Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.

```python
#WRITE YOUR CODE HERE
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
num = 5
print("Factorial of", num, "is", factorial(num))
```

5) Write a Python function to check whether a number falls within a given range.

```python
#WRITE YOUR CODE HERE
def is_in_range(num, start, end):
    return start <= num <= end
num = 10
start_range = 5
end_range = 20

if is_in_range(num, start_range, end_range):
    print(f"{num} is in the range [{start_range}, {end_range}]")
else:
    print(f"{num} is NOT in the range [{start_range}, {end_range}]")
```
